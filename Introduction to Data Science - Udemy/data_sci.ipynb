# Split Data | Verify Result | Accuracy Test

```python
from sklearn.model_selection import train_test_split
xtrain , xtest , ytrain , ytest = train_test_split(X,y,test_size = 0.3)

logistic1 = LogisticRegression(tol=0.0005, max_iter=1000)
knna1= KNeighborsClassifier(n_neighbors=1)
knna5= KNeighborsClassifier(n_neighbors=5)

knna1.fit(xtrain,ytrain)
knna5.fit(xtrain,ytrain)
logistic1.fit(xtrain,ytrain)
```

```python
ypredl=logistic.predict(xtest)

ypredk1=knna1.predict(xtest)
ypredk5=knna5.predict(xtest)
```

```python
from sklearn import metrics
print("log ",metrics.accuracy_score(ytest,ypredl))
print(metrics.accuracy_score(ytest,ypredk1))
print(metrics.accuracy_score(ytest,ypredk5))
```

# K Fold Cross Validation

```python
from sklearn.model_selection import cross_val_score

score = cross_val_score(knn5,X,y,cv=10,scoring='accuracy')
print(score.mean())
```

```python
s=[]
for i in range(1,45):
    kn= KNeighborsClassifier(n_neighbors=i)
    score = cross_val_score(kn,X,y,cv=10,scoring='accuracy')
    s.append(score.mean())
    
print(s)
```

```python
import matplotlib.pyplot as plt

plt.plot(range(1,45),s)
```
